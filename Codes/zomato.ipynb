```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('dark_background')
```

```python
# read csv file
df = pd.read_csv('zomato.csv')
df.head()
```

```python
# know rows and columns
df.shape
```

```python
df.columns
```

```python
df = df.drop(['url', 'address', 'phone', 'menu_item', 'dish_liked', 'reviews_list'] , axis=1)
df.head()
```

```python
df.info()
```

```python
# dropping duplicates
df.drop_duplicates(inplace=True)
df.shape
```

```python
# cleaning rate column
df['rate'].unique()
```

```python
# Removing "NEW" , "-" and "/5" from Rate Column
def handleRate(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)

df['rate'] = df['rate'].apply(handleRate)
df['rate'].head()
```

```python
df['rate'].unique()
```

```python
df['rate'].isnull().sum()
```

```python
# Filling Null Values in Rate Column with Mean
df['rate'].fillna(df['rate'].mean() , inplace=True)
df['rate'].isnull().sum()
```

```python
df.info()
```

```python
df.dropna(inplace = True)
df.head()
```

```python
df.info()
```

```python
# renaming names of columns 
df.rename(columns = {'approx_cost(for two people)':'Cost2plates', 
                     'listed_in(type)':'Type',
                    'listed_in(city)':'city'},
                    inplace = True)
df.head()
```

```python
df['location'].unique()
```

```python
df['city'].unique()
```

```python
# droping city , as we have location available
df = df.drop(['city'], axis = 1)
df.head()
```

```python
df['Cost2plates'].unique()
```

```python
# dropping comma from cost price
def handleComma(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',','')
        return float(value)
    else:
        return float(value)
        
df['Cost2plates'] = df['Cost2plates'].apply(handleComma)
df['Cost2plates'].unique()
```

```python
df.head()
```

```python
# cleaning rest_type column
rest_type = df['rest_type'].value_counts()
rest_type
```

```python
rest_types_lessthan1000 = rest_type[rest_type < 1000]
rest_types_lessthan1000
```

```python
# Making Rest Types less than 1000 in frequency as others

def handle_rest_type(value):
    if(value in rest_types_lessthan1000):
        return 'others'
    else:
        return value
    
df['rest_type'] = df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

```python
df.head()
```

```python
# Cleaning Location Column

location = df['location'].value_counts()
location
```

```python
location_lessthan300 = location[location<300]
location_lessthan300
```

```python
# Making loaction less than 300 in frequency as others

def handle_location(value):
    if(value in location_lessthan300):
        return 'others'
    else:
        return value
        
df['location'] = df['location'].apply(handle_location)
df['location'].value_counts()
```

```python
# Cleaning Cuisines Column

cuisines = df['cuisines'].value_counts()
cuisines_lessthan100 = cuisines[cuisines<100]
cuisines_lessthan100
```

```python
# Making cuisines less than 100 in frequency as others 

def handle_cuisines(value):
    if(value in cuisines_lessthan100):
        return 'others'
    else:
        return value
        
df['cuisines'] = df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```
